### Derivatives

The derivative of a function \( f(x) \) tells us the rate of change (slope) of the function with respect to \( x \).

For example, for \( f(x) = 3x^2 + 2x + 1 \), the derivative is:

\[
f'(x) = 6x + 2
\]

This shows how the function value changes as \( x \) changes.

---

### Partial Derivatives

When dealing with functions of multiple variables, like \( f(w_1, w_2) = 3w_1^2 + 2w_1w_2 + 4w_2^2 \), the **partial derivative** shows the rate of change with respect to **one variable at a time**, holding the others constant.

\[
\frac{\partial f}{\partial w_1} = 6w_1 + 2w_2
\]

\[
\frac{\partial f}{\partial w_2} = 2w_1 + 8w_2
\]

Partial derivatives are essential in updating parameters independently in machine learning models.


```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python

#DERIVATIVE

# Example function: f(x) = 3x^2 + 2x + 1
def f(x):
    return 3*x**2 + 2*x + 1

# Analytical derivative: f'(x) = 6x + 2
def df(x):
    return 6*x + 2

# Plot function and derivative
x = np.linspace(-3, 3, 400)
y = f(x)
dy = df(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="f(x) = 3x^2 + 2x + 1")
plt.plot(x, dy, label="f'(x) = 6x + 2", linestyle='--')
plt.title("Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PARTIAL DERIVATIVE

# Example multivariable function: f(w1, w2) = 3w1^2 + 2w1w2 + 4w2^2
def f_multi(w1, w2):
    return 3*w1**2 + 2*w1*w2 + 4*w2**2

# Partial derivatives:
# ∂f/∂w1 = 6w1 + 2w2
def df_dw1(w1, w2):
    return 6*w1 + 2*w2

# ∂f/∂w2 = 2w1 + 8w2
def df_dw2(w1, w2):
    return 2*w1 + 8*w2

# Choose a point (w1, w2)
w1_val = 1.0
w2_val = 2.0

print(f"At point w1={w1_val}, w2={w2_val}:")
print(f"Function value: f(w1, w2) = {f_multi(w1_val, w2_val)}")
print(f"Partial derivative w.r.t w1: ∂f/∂w1 = {df_dw1(w1_val, w2_val)}")
print(f"Partial derivative w.r.t w2: ∂f/∂w2 = {df_dw2(w1_val, w2_val)}")
```
