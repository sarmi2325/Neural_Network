## **LIMITS**


The concept of a limit describes the behavior of a function as its input approaches a particular value — even if the function is not explicitly defined at that point.

In machine learning, this idea parallels how optimization algorithms like gradient descent adjust weights to minimize loss: while perfect accuracy (e.g., 100%) may be theoretically unreachable, we can iteratively get arbitrarily close to the optimal solution.

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define a function
def fn(x):
  return (x**2-4)/(x-2)  # Undefined at x=2
```

```python
x=np.linspace(1.9,2.2,100)
y=fn(x)
```

This warning appears because division by zero is happening at x=2

Even though the function is undefined at x=2
, we still want to understand what value the function approaches as x gets very close to 2.

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,color='blue',label='f(x) = (x² - 4)/(x - 2)')
plt.axvline(x=2, color='red', linestyle='--', label='x = 2 (undefined)')
plt.title('Visualizing a Limit at x = 2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
