# **Continuity**

A function is continuous at a point
x = a, if there is no sudden jump, hole, or break at that point.
Formally, a function f(x) is continuous at x=a if:


x→a
lim
​
 f(x)=f(a)

This means:

The limit of f(x) as x approaches a exists.

The function f is defined at a.

The limit value and the function value at a are equal.

If any of these are false, the function is discontinuous at x=a.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Continuous function
def f_cont(x):
    return x**2
```

```python
x = np.linspace(0, 4, 400)
y_cont = f_cont(x)
```

```python
plt.figure(figsize=(12, 5))

# Continuous plot
plt.subplot(1, 2, 1)
plt.plot(x, y_cont, label='f(x) = x²', color='green')
plt.title("Continuous Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)


```

```python
def f_disc(x):
    # Function with discontinuity at x=2
    return np.where(x == 2, 10, x**2)

x = np.linspace(0, 4, 400)
y = f_disc(x)

plt.plot(x, y, label="f(x) = x² except x=2 → 10")

# Mark the discontinuity point explicitly with matching array sizes
plt.scatter([2], [4], color='red', s=100, label="Discontinuity at x=2")

plt.title("Function with Discontinuity")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
