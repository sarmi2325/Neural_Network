```python
#Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def position(seq_length,dim):
  #create an zero matrix for the final return of the positional encoding
  posi=np.zeros((seq_length,dim))

  #create an 2D array for position and dimension
  p=np.arange(seq_length)[:,np.newaxis] #(60,1)
  d=np.arange(dim)[np.newaxis,:] #(1,16)

  #calculating frequency
  freq = 1/np.power(10000,(2*(d//2))/np.float32(dim))
  rad = p*freq #(60,16)

  posi[:,0::2]=np.sin(rad[:,0::2])
  posi[:,1::2]=np.cos(rad[:,1::2])

  return posi
```

```python
seq_length=60
dim=16

positional_encoding=position(seq_length,dim)
```

```python
plt.figure(figsize=(8, 4))
for i in range(dim):
    plt.plot(np.arange(seq_length), positional_encoding[:, i], label=f'Dim {i}')
plt.title("Positional Encoding - 4 Dimensions")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
